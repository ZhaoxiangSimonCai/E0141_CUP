```python
import pandas as pd
```

```python
histo_df = pd.read_excel(
    "/home/scai/scratch/E0141_CUP/data/E0141_histo.xlsx"
)
prolims_df = pd.read_excel(
    "/home/scai/scratch/E0141_CUP/data/E0141_prolims.xlsx"
)

meta_df = pd.read_excel(
    "/home/scai/scratch/E0141_CUP/data/E0141_meta.xlsx"
)
```

```python
meta_df
```

```python
combined_meta_df = pd.merge(
    prolims_df, histo_df[["Subsample LIMS ID", "Image Label"]], on="Subsample LIMS ID"
)
combined_meta_df = pd.merge(
    combined_meta_df, meta_df, left_on="Prep LIMS ID", right_on="prep_lims_id", how="left"
)
combined_meta_df.drop(columns=["MS Filename"], inplace=True)
combined_meta_df.drop_duplicates(subset=["Prep LIMS ID"], inplace=True)
```

```python
combined_meta_df.set_index("Prep LIMS ID", inplace=True)
combined_meta_df.to_excel(
    "/home/scai/scratch/E0141_CUP/data/E0141_combined_meta.xlsx"
)
```
